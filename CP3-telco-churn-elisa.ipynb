## **A. Business Problem Understanding**

### **a. Context**

Industri telekomunikasi merupakan salah satu industri yang sangat kompetitif, dengan berbagai _service provider_ yang menawarkan paket dan fitur yang berbeda-beda untuk menarik dan mempertahankan _customer_. Oleh karena itu, mengurangi _churn_ bukan hanya tentang mempertahankan jumlah pelanggan, tetapi juga tentang mengoptimalkan profitabilitas. Masalah yang perlu diatasi adalah mengidentifikasi faktor-faktor yang berkontribusi terhadap _customer churn_ dan memprediksi _customer_ mana yang kemungkinan besar akan meninggalkan atau berhenti berlangganan. ### **e. Metric Evaluation**

Dari permasalahan dan tujuan yang telah dijelaskan di atas, maka fokus utama kita adalah memprediksi *customer* yang lebih mungkin untuk *churn*, maka target yang kita tetapkan sebagai berikut.

Target :
- 0 : Tidak berhenti berlangganan
- 1 : Berhenti berlangganan (churn)

Type 1 error : False Positive (*customer* yang aktualnya tidak churn tetapi diprediksi churn)
Konsekuensi: tidak efektifnya pemberian insentif retensi

Type 2 error : False Negative (*customer* yang aktualnya churn tetapi diprediksi tidak akan churn)
Konsekuensi: kehilangan *customer*

Dengan kategorisasi di atas, sekarang kita dapat memilih apa yang penting dalam bisnis yang ingin memprioritaskan tingkat retensi. Dalam hal ini, kita harus fokus pada **meminimalkan jumlah False Negative** dalam prediksi model kita. 

Pada permasalah ini, lebih baik salah memprediksi *customer* yang akan churn tetapi sebenarnya mereka tidak, daripada salah memprediksi *customer* yang tidak akan churn tetapi sebenarnya mereka churn.

Dalam skenario terburuk pertama, jika kita salah memprediksi bahwa mereka akan churn (type 1 error), kemungkinan perusahaan harus mengeluarkan lebih banyak biaya pemasaran pada *customer* yang sudah ada.

Di sisi lain, skenario kedua (type 2 error) akan membuat perusahaan harus mengeluarkan lebih banyak pengeluaran untuk mendapatkan *customer* baru. Karena model akan salah memprediksi *customer* yang tidak akan churn tetapi sebenarnya mereka churn, perusahaan akan kehilangan kesempatan untuk memaksimalkan *customer* yang sudah ada dan pada akhirnya perlu mengeluarkan biaya lebih banyak untuk mendapatkan *customer* baru.

Untuk gambaran secara kuantitatif, kita dapat mencoba untuk hitung dampak biaya berdasarkan asumsi berikut:

- Customer Living Period untuk *customer* yang churn sekitar 17 bulan.
- Customer Acquisition Cost (CAC) = $315 per customer ([sumber](https://www.revechat.com/blog/customer-acquisition-cost/)) / 17 bulan = $18.53 per bulan per customer
- Customer Retention Cost (CRC)= 1/5 ([sumber](https://www.outboundengine.com/blog/customer-retention-marketing-vs-customer-acquisition-marketing/)) * CAC = 1/5 * $18.53 = $3.7 per bulan per customer
- Average Customer MonthlyCharge = $64.88 per bulan per customer

Berdasarkan asumsi di atas maka kita dapat coba kuantifikasi konsekuensinya sebagai berikut:
- Tidak efektifnya pemberian insentif retensi, maka kita **menyia-nyiakan biaya CRC sebesar $3.7 per bulan per *customer***
- Kehilangan *customer*, maka kita **kehilangan pendapatan dan juga perlu mengeluarkan kembali biaya CAC** sehingga <u>secara total kita kehilangan $18.53 + $64.88 = $83.41 per bulan per *customer*</u>

Berdasarkan konsekuensinya, maka sebisa mungkin yang akan kita lakukan adalah membuat model yang dapat mengurangi *customer churn* dari perusahaan tersebut, khususnya jumlah **False Negative** (*customer* yang aktualnya churn tetapi diprediksi tidak akan churn), tetapi juga dapat meminimalisir pemberian insentif yang tidak tepat. Jadi nanti metric utama yang akan kita gunakan adalah `f2_score`, karena **recall kita anggap dua kali lebih penting daripada precision**.

Pada data numerikal, dilakukan uji normalitas untuk mengidentifikasi apakah data terdistribusi normal atau tidak. Dipilih **Shapiro-Wilk Distribution test** untuk menguji normalitas karena jumlah data kurang dari 5000. Uji ini memberikan nilai p (p-value), yang menunjukkan apakah data berbeda signifikan dari distribusi normal.

**Interpretasi**:
- `p-value > 0.05`: Fail to Reject H0. Tidak ada cukup bukti untuk menolak hipotesis nol (data berdistribusi normal).
- `p-value â‰¤ 0.05`: Reject H0. Ada bukti untuk menolak hipotesis nol (data tidak berdistribusi normal).

```python
from scipy.stats import shapiro

dist = []

for i in range(len(numerical)):
    feature_name = numerical[i]
    stat, p_value = shapiro(data[feature_name].dropna())
    
    if p_value > 0.05:
        distribution = "Normally Distributed"
    else:
        distribution = "Not Normally Distributed"
    
    dist.append([feature_name, stat, p_value, distribution])

dist_df = pd.DataFrame(dist, columns=['Feature', "Shapiro", 'P-value', 'Distributed'])
dist_df
```

    - `MonthlyCharges`:

> **Chi-square test**

Uji Chi-Square digunakan untuk menguji independensi antara dua variabel kategorikal. Dalam konteks ini, uji Chi-Square digunakan untuk menentukan apakah ada hubungan signifikan antara churn dan setiap fitur kategorikal (seperti Dependents, OnlineSecurity, InternetService, dll.).

```python
categorical
```

```python
from scipy.stats import chi2_contingency

def chi2_calculator(col1, col2):
    alpha = 0.05
    contingency_table = pd.crosstab(col1, col2)
    stat, pvalue, dof, expected_freq = chi2_contingency(contingency_table)
    if pvalue > alpha:
        result = "Fail to Reject Ho"
        judgement = "Independent"
    else:
        result = "Reject Ho"
        judgement = "Dependent"
    return result, pvalue, judgement

results = []
judgement = []

plot_columns = categorical

for i in plot_columns:
    result, pvalue,judgement = chi2_calculator(data['Churn'], data[i])
    results.append({'column': i, 'pvalue': pvalue, 'result': result,'verdict': judgement})

results_df = pd.DataFrame(results)
results_df
```

<u>**Insights**</u>:

- Semua feature yang diuji menunjukkan **p-value yang sangat kecil (jauh lebih kecil dari 0.05)**. Hal ini ditunjukkan oleh nilai p-value yang sangat rendah seperti 4.700746e-33 untuk `Dependents`, 7.488416e-127 untuk `OnlineSecurity`, dan sebagainya.

### **c. Outliers**

Kita sudah cek bahwa secara umum, data pada kolom numerikal `tenure` dan `MonthlyCharges` tidak terdapat outlier.

Hal ini juga dapat dilihat pada tabel di bawah, di mana seluruh data pada feature `tenure` dan `MonthlyCharges` berada di dalam rentang ***lower bound*** (Q1 - 1.5 IQR) dan ***upper bound*** (Q3 + 1.5 IQR).

```python
def calculate_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

outlier_data = []

numerical = [var for var in data.columns if data[var].dtypes != 'object']

for column in numerical:
    outlier_result, lower_bound, upper_bound = calculate_outliers(data, column)
    total_outlier = len(outlier_result)
    outlier_percentage = round(total_outlier / len(data[column]) * 100, 2)
    outlier_data.append([column, total_outlier, outlier_percentage, lower_bound, upper_bound])

outlier_df = pd.DataFrame(outlier_data, columns=["Column", "Total Outliers", "Percentage (%)", "Lower Bound", "Upper Bound"])
outlier_df
```

### **g. Rare Values**

Pada tahapan ini dilakukan pengecekan apakah terdapat *rare values* dari tiap feature. *Threshold* yang digunakan adalah 5%. Hal ini bertujuan untuk menentukan sebuah kategori atau nilai dianggap *rare values* jika muncul **kurang dari 5% dari total data**.

```python
def find_rare_values(data, categorical, threshold):
    rare_values = {}
    for col in categorical:
        value_counts = data[col].value_counts()
        rare = value_counts[value_counts < threshold * len(data)]
        if not rare.empty:
            rare_values[col] = rare
    return rare_values

# Define a threshold for rare values (Here we choose 5%)
threshold = 0.05

# Find rare values
rare_values = find_rare_values(data, categorical, threshold)

for col in categorical:
    if col in rare_values:
        print(f"Rare values in {col}:")
        print(rare_values[col])
    else:
        print(f"No rare values in {col}")
```

```python
def list_feature_values(data, categorical):
    feature_values_list = []
    for col in categorical:
        value_counts = data[col].value_counts()
        for value, count in value_counts.items():
            percentage = count / len(data) * 100
            feature_values_list.append([col, value, count, f"{percentage:.2f}%"])
    return feature_values_list

feature_values_list = list_feature_values(data, categorical)
feature_values_data = pd.DataFrame(feature_values_list, columns=['Feature', 'Value', 'Count', 'Percentage'])

feature_values_data
```

Dalam dataset ini, tidak terdapat *rare values* pada setiap feature yang ada. Data pada dataset ini *imbalance*, tetapi tidak ada data pada suatu kategori yang berjumlah kurang dari 5% dari total dataset.

```python
def create_plot(data, col, x_maks=1000):

    agg_table = data.groupby([col, 'Churn']).size().unstack().reset_index()
    agg_table['total'] = agg_table['Yes']+agg_table['No']
    agg_table['% Churn'] = (agg_table['Yes']/agg_table['total']*100).round(2)
    agg_table = agg_table.sort_values(by='% Churn', ascending=False)
    agg_table

    fig, axes = plt.subplots(1, 2, figsize=(16,5))

    ax1 = sns.barplot(data=agg_table, y=col, x='% Churn', ax=axes[0])
    label = ['{:.1f} %'.format(x) for x in ax1.containers[0].datavalues]
    ax1.bar_label(ax1.containers[0], labels=label, padding=3, label_type='center')
    ax1.set_title(f'Percentage of Churn Customer by {col}')

    ax2 = sns.countplot(data=data, y=col, hue='Churn', order=agg_table[col], ax=axes[1])
    ax2.set_title(f'Number of customer by {col} and churn')
    for container in ax2.containers:
        ax2.bar_label(container)
    ax2.set_xlim(0, x_maks)

    plt.tight_layout()
    plt.show()
```

### **d. Pipeline**

Proses proprocessing dimasukkan ke dalam pipeline.

```python
# Encoding
transformer = ColumnTransformer([
    ('onehot', OneHotEncoder(), ['Dependents', 'OnlineSecurity', 'OnlineBackup',
       'InternetService', 'DeviceProtection', 'TechSupport', 'Contract',
       'PaperlessBilling'])
], remainder='passthrough')
```

```python
# Scaling
scaler = MinMaxScaler()
```

```python
data.columns
```

```python
# Resampling
resample = SMOTENC(random_state=0, categorical_features=[0, 2, 3, 4, 5, 6, 7, 8]) #['Dependents', 'OnlineSecurity', 'OnlineBackup', 'InternetService', 
                                                                                    # 'DeviceProtection', 'TechSupport', 'Contract', 'PaperlessBilling']
```

```python
# Resampling untuk hyperparameter tuning
smote = SMOTE(random_state=0)
rus = RandomUnderSampler(random_state=0)
ros = RandomOverSampler(random_state=0)

resamplers = [rus, ros, smote, resample]
```

```python
Pipeline([
    ('transformer', transformer),
    ('scaler', scaler),
    ('resampler', resample),
])
```

### **e. Model Benchmarking**

```python
# Define the models

# Logistic Regression 
logreg = LogisticRegression(random_state=0)

# KNN
knn = KNeighborsClassifier()

# Decision Tree
tree = DecisionTreeClassifier(random_state=0)

# Base Models
logreg_pipeline = Pipeline([
    ('scaling', scaler),
    ('modeling', logreg)
])

knn_pipeline = Pipeline([
    ('scaling', scaler),
    ('modeling', knn)
])

dtree_pipeline = Pipeline([
    ('modeling', tree)
])

# Meta learner
meta_logreg = LogisticRegression(random_state=0)

# Voting Classifier (Hard)
voting_hard = VotingClassifier(
    estimators=[
        ('clf1', logreg_pipeline),
        ('clf2', knn_pipeline),
        ('clf3', dtree_pipeline)
    ], voting='hard'
)

# Voting Classifier (Soft)
voting_soft = VotingClassifier(
    estimators=[
        ('clf1', logreg_pipeline),
        ('clf2', knn_pipeline),
        ('clf3', dtree_pipeline)
    ], voting='soft'
)

# Stacking Classifier
stacking = StackingClassifier(
    estimators=[
        ('clf1', logreg_pipeline),
        ('clf2', knn_pipeline),
        ('clf3', dtree_pipeline)
    ],
    final_estimator=meta_logreg
)

# Bagging Classifier
bagging_clf = BaggingClassifier(
    estimator=logreg,
    random_state=0
)

# CatBoost
cat = CatBoostClassifier(random_state=0, verbose=False)

# LightGBM
lgbm = LGBMClassifier(random_state=0)

# RandomForest
rf = RandomForestClassifier(random_state=0)

# Adaboost
adaboost = AdaBoostClassifier(random_state=0)

# Gradientboost
gradboost = GradientBoostingClassifier(random_state=0)

# ExtremeGradientBoost
xgboost = XGBClassifier(random_state=0, verbosity=0)
```

```python
from sklearn.metrics import fbeta_score, make_scorer

# Create scorer
f2_score = make_scorer(fbeta_score, beta=2)
```

```python
# Pilihan model
models = [logreg, knn, tree, voting_hard, voting_soft, stacking, bagging_clf, cat, lgbm, rf, adaboost, gradboost, xgboost]
model_names = ['LogisticRegression', 'KNeigborsClassifier', 'DecisionTree',
               'Voting Classifier (Hard)', 'Voting Classifier (Soft)', 'Stacking Classifier',
               'BaggingClassifier', 'CatBoostClassifier', 'LightGBMClassifier', 'RandomForestClassifier', 'AdaBoostClassifier', 
               'GradientBoostingClassifier', 'XGBClassifier']

f2_mean_resampler = []
f2_std_resampler = []
all_f2_resampler = []

f2_mean = []
f2_std = []
all_f2 = []


# Mencari model terbaik berdasarkan rerata f2 score terkecil dan std terkecil
for algoritma in models:

    # dengan resampler
    model_pipe = Pipeline([
        ('transformer', transformer),
        ('scaler', scaler),
        ('resampler', resample),
        ('model', algoritma)
    ])

    skfold = StratifiedKFold(n_splits = 5)

    model_cv = cross_val_score(
        model_pipe,                 # model yang sudah dimasukkan ke dalam pipeline
        X_train,                    # data sebelum di preprocessing
        y_train, 
        cv = skfold, 
        scoring = f2_score,
        error_score='raise'
        )

    f2_mean_resampler.append(model_cv.mean())
    f2_std_resampler.append(model_cv.std())
    all_f2_resampler.append(model_cv.round(4))

    # tanpa resampler
    model_pipe = Pipeline([
        ('transformer', transformer),
        ('scaler', scaler),
        ('model', algoritma)
    ])

    skfold = StratifiedKFold(n_splits = 5)

    model_cv = cross_val_score(
        model_pipe,                 # model yang sudah dimasukkan ke dalam pipeline
        X_train,                    # data sebelum di preprocessing
        y_train, 
        cv = skfold, 
        scoring = f2_score,
        error_score='raise'
        )

    f2_mean.append(model_cv.mean())
    f2_std.append(model_cv.std())
    all_f2.append(model_cv.round(4))
```

```python
# Perbandingan f2 score pada tiap algoritma
pd.DataFrame({
    'model': model_names,
    'f2 score mean with resampler': f2_mean_resampler,
    'std with resampler': f2_std_resampler,
    'all with score resampler' : all_f2_resampler,
    'f2 score mean without resampler': f2_mean,
    'std without resampler': f2_std,
    'all without score resampler' : all_f2
}).sort_values('f2 score mean with resampler', ascending=False)
```

<u>**Insights**</u>:

- Model AdaBoostClassifier menunjukkan kinerja terbaik dengan F2 score rata-rata 0.718020 dengan resampler dan menggunakan default hyperparameter, diikuti oleh GradientBoostingClassifier dengan F2 score 0.701508 dan LogisticRegression dengan F2 score 0.698272. Nilai standar deviasi (std) untuk AdaBoostClassifier adalah 0.017594, yang relatif rendah, menunjukkan kinerja model yang cukup konsisten.
- Sebagian besar model menunjukkan peningkatan signifikan dalam F2 score saat menggunakan resampler, terlihat dari perbedaan hasil scoring pada kolom "f2 score mean with resampler" dan "f2 score mean without resampler". Seperti pada model AdaBoostClassifier, yang mengalami peningkatan F2 score dari 0.541092 tanpa resampler menjadi 0.718020 dengan resampler, menunjukkan bahwa resampling membantu model ini mengatasi ketidakseimbangan data atau meningkatkan performa secara keseluruhan.

Untuk mengukur performa akhir model

```python
scores_train = []
scores_test = []

# Mencari model terbaik berdasarkan balanced_accuracy score terbesar dan std terkecil
def y_pred_func(model):
    model_pipe=Pipeline([
        ('preprocessing', transformer), 
        ('scaler', scaler), 
        ('smote', resample),
        ('model', model)
    ])

    # fitting
    model_pipe.fit(X_train, y_train)

    return model_pipe

for algoritma in models:
    model_pipe = y_pred_func(algoritma)

    # predict
    y_pred_train = model_pipe.predict(X_train)
    y_pred_test = model_pipe.predict(X_test)

    # evaluate
    score_train = fbeta_score(y_train, y_pred_train, beta=2)
    score_test = fbeta_score(y_test, y_pred_test, beta=2)
    
    scores_train.append(score_train)
    scores_test.append(score_test)
```

### **f. Hyperparameter Tuning**

> #### **Hyperparameter Tuning: AdaBoostClassifier**

```python
# learning rate
learning_rate = list(np.arange(0.01,1, 0.01))

# jumlah pohon
n_estimators = list(np.arange(50,500))

# algoritma untuk mencari AdaBoost terbaik
algorithm = ['SAMME', 'SAMME.R']

# Hyperparameter space Adaboost
hyperparam_space_adaboost = {
    'resampler':[smote, rus, ros, resample, None],
    'model__learning_rate':learning_rate,
    'model__n_estimators':n_estimators,
    'model__algorithm':algorithm,
}
```

```python
# Benchmark model 1
adaboost = AdaBoostClassifier(random_state=0)

# Stratified cross validation
skfold = StratifiedKFold(n_splits = 5)

# Create the algorithm chains
estimator_adaboost = Pipeline([
    ('transformer', transformer),
    ('scaler', scaler),
    ('resampler', resample),
    ('model', adaboost)
    ])

# Hyperparameter tuning
tuned_adaboost = RandomizedSearchCV(
    estimator=estimator_adaboost, 
    param_distributions = hyperparam_space_adaboost, 
    cv = skfold, 
    scoring = f2_score, 
    n_jobs = -1, 
    random_state=0,
    n_iter=500
)
```

```python
# fitting 
tuned_adaboost.fit(X_train, y_train)
```

```python
# # score terbaik hasil dari randomizedSearch
print(f'F2 Score pada train set: {tuned_adaboost.best_score_:.3f}')

# # hyperparameter pada score terbaik hasil dari tuned_adaboost
print(f'Best parameter: {tuned_adaboost.best_params_}')
```

### 1. **Recall (Sensitivity or True Positive Rate)**
Recall measures the proportion of actual positive cases that are correctly identified by the model.

$$\text{Recall} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Negatives (FN)}}\$$

### 2. **Precision (Positive Predictive Value)**
Precision measures the proportion of predicted positive cases that are correctly actual positives.

$$\text{Precision} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Positives (FP)}}\$$

### 3. **F2-Score**
The F2-score is a metric that combines precision and recall but gives more weight to recall.

$$\text{F2-Score} = \left(1 + 2^2\right) \cdot \frac{\text{Precision} \times \text{Recall}}{(2^2 \times \text{Precision}) + \text{Recall}}\$$

Simplified:

$$\text{F2-Score} = \frac{5 \times \text{Precision} \times \text{Recall}}{4 \times \text{Precision} + \text{Recall}}\$$

### 4. **Accuracy**
Accuracy measures the proportion of all cases (both positives and negatives) that are correctly identified.

$$\text{Accuracy} = \frac{\text{True Positives (TP)} + \text{True Negatives (TN)}}{\text{True Positives (TP)} + \text{True Negatives (TN)} + \text{False Positives (FP)} + \text{False Negatives (FN)}}\$$

| Precision | 0.4661 | Dari keseluruhan *customer* yang diklasifikasikan ***churn*** sebanyak 445 orang, **hanya 46.61% yang diprediksi dengan benar** (234 orang) |
| F2-Score | 0.7627 | Nilai F2-Score didapatkan dari (5 * Precison * Recall)/(4 * Precision + Recall) = (5 * 46.61% * 90.69%) / (4 * 46.61% + 90.69%) = **76.27%** |
| Accuracy |  0.6992 | Dari keseluruhan *customer* yang dites sebanyak 971 orang, **69.92% data diprediksi dengan benar** (678 orang) |

    - `Total biaya FP`: $3.7 * 268 = `$991.6 per bulan`.

    - TP adalah ***customer* yang benar-benar churn dan diprediksi dengan benar untuk churn**. Biaya retensi yang dihabiskan pada *customer* ini dianggap **efektif**.
    - `Total biaya TP`: $3.7 * 234 = `$865.8 per bulan`.

- Keuntungan dari Retensi True Negative (TN):
    - TN adalah *customer* yang tidak churn dan diprediksi dengan benar untuk tidak churn, tidak ada biaya tambahan.

Kesimpulan: `Total Biaya Keseluruhan`: $2,001.84 (FN) + $991.6 (FP) + $865.8 (TP) = `$3,859.24 per bulan`.

Semua customer yang churn tidak akan terdeteksi sebelumnya.     - `Total biaya churn tanpa prediksi`: $83.41 * 234 = `$19,517.94 per bulan`.

- Biaya Retensi yang Tidak Efektif (Tanpa Prediksi):

    Semua customer akan mendapatkan perlakuan retensi umum tanpa prediksi, maka biaya ini **lebih tinggi karena tidak efisien**. Misalkan perusahaan melakukan retensi pada seluruh customer (971 customer), tanpa memprediksi siapa yang akan churn:

`Total biaya tanpa prediksi`: $19,517.94 (churn) + $3,592.7 (retensi) = `$23,110.64 per bulan.`
